### 10/6일 금요일

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

#행정동 코드(서원, 신원, 서림, 신림동) 
hdong_cd_list = [1162064500, 1162065500, 1162066500, 1162069500]
```

```python
#23년 9월 1일, 8일, 15일, 22일, 29일 
df_sep_1 = pd.read_csv("csv\stay_20230917_1.csv")
df_sep_8 = pd.read_csv(r"D:\빅데이터 공모전 관련 자료\체류 데이터 파일\stay_20230908_1.csv")
df_sep_15 = pd.read_csv(r"D:\빅데이터 공모전 관련 자료\체류 데이터 파일\stay_20230915_1.csv")
df_sep_22 = pd.read_csv(r"D:\빅데이터 공모전 관련 자료\체류 데이터 파일\stay_20230922_1.csv")
df_sep_29 = pd.read_csv(r"D:\빅데이터 공모전 관련 자료\체류 데이터 파일\stay_20230929_1.csv")
```

```python
#각 일자에 따른 지역구 구분화 
filtered_data_1 = df_sep_1[df_sep_1['hdong_cd'].isin(hdong_cd_list)]
filtered_data_8 = df_sep_8[df_sep_8['hdong_cd'].isin(hdong_cd_list)]
filtered_data_15 = df_sep_15[df_sep_15['hdong_cd'].isin(hdong_cd_list)]
filtered_data_22 = df_sep_22[df_sep_22['hdong_cd'].isin(hdong_cd_list)]
filtered_data_29 = df_sep_29[df_sep_29['hdong_cd'].isin(hdong_cd_list)]
```

```python
#각 일자에 따른 지역구 합
f_1_sum=filtered_data_1.groupby('time')['stay_cnts'].sum()
f_8_sum=filtered_data_8.groupby('time')['stay_cnts'].sum()
f_15_sum=filtered_data_15.groupby('time')['stay_cnts'].sum()
f_22_sum=filtered_data_22.groupby('time')['stay_cnts'].sum()
f_29_sum=filtered_data_29.groupby('time')['stay_cnts'].sum()
```

```python
#지역구에 따른 평균 
f_sum = [f_1_sum, f_8_sum, f_15_sum, f_22_sum, f_29_sum]
avg_df = pd.concat(f_sum).groupby('time').mean()
print(avg_df)
```

```python
#23년 10월 6일 
df_oct_6 = pd.read_csv(r"D:\빅데이터 공모전 관련 자료\체류 데이터 파일\stay_20231006_1.csv")
filtered_data_6 = df_oct_6[df_oct_6['hdong_cd'].isin(hdong_cd_list)]
f_6_sum=filtered_data_6.groupby('time')['stay_cnts'].sum()
print(f_6_sum)
```

```python
# 한글 폰트 설정 (malgun 폰트 사용 예시)
font_path = "C:/Windows/Fonts/malgun.ttf"  
font_prop = fm.FontProperties(fname=font_path)
plt.rc('font', family=font_prop.get_name())
```

```python
# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(avg_df.index, avg_df.values, color='blue', label='평상시 체류인구')
plt.plot(f_6_sum.index, f_6_sum.values, color='red', label='축제시 체류인구')
# 그래프 설정
plt.title('평상시 vs 축제시 체류인구 수')
plt.xlabel('시간 (time)')
plt.ylabel('체류인구 수 (stay_cnts)')
plt.legend()
plt.grid(True)

# 그래프 보여주기
plt.show()
```

```python
# 두 집단의 차이가 있는지 
import numpy as np
from scipy import stats

group_com = avg_df.values
group_fes = f_6_sum.values
print(stats.ttest_ind(group_com, group_fes))
print(stats.ttest_ind(group_com, group_fes,equal_var=False))
```
